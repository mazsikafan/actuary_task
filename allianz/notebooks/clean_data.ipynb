```python
import pandas as pd
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
```

This notebook contains the code to merge and perform preliminary cleaning on the excel files downloaded from https://akk.hu/statistics/yields-indices-market-turnover/benchmark-yields
This data was choosen in order to have daily data for the Hungarian government bond yields, which is not available in the Hungarian central bank's database.

```python
path = '/Users/balintmaroti/Documents/allianz/data'
files = glob.glob(os.path.join(path, "benchmark_*.xlsx"))
files.sort()
```

```python
dfs = []
for fn in files:
    df = pd.read_excel(fn, sheet_name="Sheet1")
    df["source_file"] = os.path.basename(fn)
    dfs.append(df)
    
full = pd.concat(dfs, ignore_index=True)
print(f"Combined shape: {full.shape}")
```

```python
full
```

```python
full.dtypes
```

```python
full.columns = (
    full.columns
        .str.strip()
        .str.lower()
        .str.replace(r'[\s%()]+', '_', regex=True)  # spaces, %, (), → underscore - ChatGPT
        .str.strip('_')
)
full
```

```python


# Replace NaN values in 'change_pct' with 0
full['change'].fillna(0, inplace=True)

full
```

```python
# check for duplicates in date and tenor and print the count
duplicates = full.duplicated(subset=['date', 'tenor'], keep=False)
print(f"Number of duplicate rows based on 'date' and 'tenor': {duplicates.sum()}")
```

```python
tenor_mapping = {
    '3 month': '3M',
    '6 month': '6M',
    '12 month': '1Y',
    '3 year': '3Y',
    '5 year': '5Y',
    '10 year': '10Y',
    '15 year': '15Y',
    '20 year': '20Y'
}

# Apply the renaming to the 'tenor' column
full['tenor'] = full['tenor'].replace(tenor_mapping)
```

```python
# Save as a csv
output_path = os.path.join(path, "benchmark_combined.csv")
full.to_csv(output_path, index=False)
```
